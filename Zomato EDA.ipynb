```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#now combining both country code and how says that left join right join in left we have more columns (focus is more)
pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
country_name=final_df.Country.value_counts().index
country_name
```

```python
country_value=final_df.Country.value_counts().values
country_value
```

```python
#pie chart to visualize-Top 3 countries uses zomato 
plt.pie(country_value[:3],labels=country_name[:3],autopct='%1.2f%%') #autopecent
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
#0  has 2148 rating counts 1 has 1 rating count 
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
```

```python
ratings
```

### Observations 
Many people did not rated the product, not rated was 2148 counts

```python
ratings.head(2)
```

```python
#plot
plt.rcParams['figure.figsize']=(12,6) #code of matlplotlib 
sns.barplot(x="Aggregate rating",y="Rating Counts",data=ratings)
```

```python
#now changing the rating color to view clearly 
sns.barplot(x="Aggregate rating",y="Rating Counts",hue='Rating color',data=ratings,palette=['brown','red','orange','yellow','green','green'])
```

### Observations 
1.Many people did not rated the product, not rated was 2148 counts 

2.The count is high and maximum number of ratings is between 2.5 to 3.4 

```python
#how frequently it is appearing 
sns.countplot(x='Rating color',data=ratings,palette=['brown','red','orange','yellow','green','green'])
```
